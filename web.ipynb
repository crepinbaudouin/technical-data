```python
import csv
import requests
from bs4 import BeautifulSoup

# Création du fichier CSV
csv_file = open('Baseconcbe.csv', 'w', encoding='utf-8', newline='')
fieldnames = ['Marque', 'adresse']
writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
writer.writeheader()

# Parcours de toutes les pages
for page in range(1, 192):
    url = f'https://www.carburants.org/station/liste/prix-a-jour/?p={page}'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    
    for station in soup.find_all('div', {'class': 'd-block'}):
        name_elem = station.find('a')
        address_elem = station.find('small')
        if name_elem is not None and address_elem is not None:
            name = name_elem.text.strip()
            address = address_elem.text.strip()
            writer.writerow({'Marque': name, 'adresse': address})
    
    print(f'Page {page} traitée')
    
csv_file.close()
print('Terminé')
```

```python
#Encodage du fichier html.parser

import requests
from bs4 import BeautifulSoup
import csv

# Liste des numéros de département à traiter
departements = [str(i).zfill(2) for i in range(1, 96)]

# Fonction pour extraire les informations de chaque station-service
def extract_station_info(station):
    id = station.get('data-station-id')
    nom = station.select_one('.info a').text
    emplacement = station.select_one('.info span').text
    return [id, nom, emplacement]

# Parcourir tous les départements de la liste et extraire les stations-service
with open('stationbe.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['id', 'nom', 'emplacement'])
    
    for dep in departements:
        url = f'https://lepompiste.fr/station-service/{dep}-departement'
        response = requests.get(url)
        soup = BeautifulSoup(response.content, 'html.parser')
        stations = soup.select('#list .data')
        
        for station in stations:
            station_info = extract_station_info(station)
            writer.writerow(station_info)
```
